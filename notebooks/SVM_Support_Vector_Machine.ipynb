```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
# Generate a simple linearly separable dataset
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
# Create an SVM classifier with a linear kernel
svm_classifier = SVC(kernel='linear', C=1.0)
# Train the classifier
svm_classifier.fit(X, y)
# Get the separating hyperplane
w = svm_classifier.coef_[0]
b = svm_classifier.intercept_[0]
x_plot = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_plot = -(w[0] / w[1]) * x_plot - b / w[1]
# Get the margins
margin = 1 / np.sqrt(np.sum(svm_classifier.coef_ ** 2))
y_margin_pos = y_plot + np.sqrt(1 + (w[0] / w[1]) ** 2) * margin
y_margin_neg = y_plot - np.sqrt(1 + (w[0] / w[1]) ** 2) * margin
# Plot the dataset, decision boundary, margins, and support vectors
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='autumn')
plt.plot(x_plot, y_plot, 'k-', label='Decision Boundary')
plt.plot(x_plot, y_margin_pos, 'k--', label='Margin')
plt.plot(x_plot, y_margin_neg, 'k--')
# Plot support vectors
plt.scatter(svm_classifier.support_vectors_[:, 0], svm_classifier.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundary and Margins')
plt.show()
```

```python
x, y = datasets.make_blobs(n_samples = 100, centers = 2, random_state = 0)
```

```python
svm_classifier = SVC(kernel = "poly", degree = 3, C = 1.0)
```

```python
svm_classifier.fit(x, y)
```

```python
xmin, xmax = x[:, 0].min() - 1, x[:, 0].max() + 1
```

```python
ymin, ymax = x[:, 1].min() - 1, x[:, 1].max() + 1
```

```python
xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100))
```

```python
z = svm_classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
```

```python
z = z.reshape(xx.shape)
```

```python
fig = plt.figure()

axis = fig.add_subplot(111, projection = "3d")
axis.plot_surface(xx, yy, z, cmap = "mako", alpha = 0.75)
scatter = axis.scatter(x[:, 0], x[:, 1], c = y)
support_vectors = svm_classifier.support_vectors_

axis.scatter(support_vectors[:, 0], support_vectors[:, 1], svm_classifier.decision_function(support_vectors), s = 100, edgecolors = "k", label = "Support Vectors")
axis.set_xlabel("Feature One")
axis.set_ylabel("Feature Two")
axis.set_zlabel("Decision Function Value")
axis.set_title("SVM decision boundary with Polynomial Kernel")

plt.show()
```
