# ***Bayes Theorem***


Bayes Theorem
*   P(A/B) = (P(B/A) . P(A)) / P(B)
*   P(A/B) = P(AnB) / P(B) ------> from equation one
*   equation one can also be written as --> P(B/A) = P(BnA) / P(A)
*   P(B/A) = P(BnA) / P(A)
*   Multiply both sides with P(A)
*   P(A) . P(B/A) = (P(AnB) . P(A)) / P(A)
*   P(A) . P(B/A) = P(AnB)
*   P(AnB) can be written as P(A/B) . P(B)
*   P(A) . P(B/A) = P(A/B) . P(B)



---------------------------------------------



*   From above Equation we get the Bayes Theorem
*   **P(A/B) = (P(A) . P(B/A)) / P(B)**

Binary classification

```python
df1 = pd.read_csv("/content/hearing_test1.csv")
```

```python
df1
```

```python
x = df1.iloc[:,0:2].values
y = df1.iloc[:,2].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
g = GaussianNB()
```

```python
y_train = y_train.ravel()

g.fit(x_train,y_train)
```

```python
y_pred = g.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
