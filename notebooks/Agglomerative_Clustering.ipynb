# Mall Customers Implementation

**Implementation Of Agglomerative Clustering**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.cluster.hierarchy as hc
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
df
```

```python
df.drop("CustomerID", axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
x = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

**Single Linkage**

```python
plt.figure(figsize = (10, 10))
dendogram = hc.dendrogram(hc.linkage(x, method = "single"))

plt.xlabel("Mall Customers")
plt.ylabel("Euclidean Distances")
plt.title("Mall Customers Single Method Dendogram")

plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters = 5, metric = "euclidean", linkage = "single")
```

```python
y_hc = ac.fit_predict(x)
```

**Maximum Linkage**

```python
plt.figure(figsize = (10, 10))
dendogram = hc.dendrogram(hc.linkage(x, method = "complete"))

plt.xlabel("Mall Customers")
plt.ylabel("Euclidean Distances")
plt.title("Mall Customers Complete Method Dendogram")

plt.show()
```

**Average Linkage**

```python
plt.figure(figsize = (10, 10))
dendogram = hc.dendrogram(hc.linkage(x, method = "average"))

plt.xlabel("Mall Customers")
plt.ylabel("Euclidean Distances")
plt.title("Mall Customers Average Method Dendogram")

plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters = 5, metric = "euclidean", linkage = "single")
```

```python
# A = (1, 2)
# B = (3, 4)
# C = (5, 6)
# D = (8, 8)
# Step 1: Calculate the initial proximity matrix
# D(A, B) = 2.83
# D(A, C) = 5.66
# D(A, D) = 9.22
# D(B, C) = 2.83
# D(B, D) = 6.40
# D(C, D) = 3.61

#     A     B     C     D
#  A  0
#  B  2.83  0
#  C  5.66  2.83  0
#  D  9.22  6.40  3.61  0
# Step 2: Compute inital ward linkage distances
# D(Ci, Cj) = Ni*Nj/(Ni+Nj)*(Meani-Meanj)**2
# Ni, Nj are the # of points in clusters Ci, Cj
# Meani, Meanj are the centroids of clusters Ci, Cj
# (Meani-Meanj)**2 is the squared Euclidean distance between the centroids
# Initially Ni, Nj are equal to 1
# D(Ci, Cj) = (xi-xj)**2

# Squared distances are:
# D(A, B)**2 = 8
# D(A, C)**2 = 32
# D(A, D)**2 = 85
# D(B, C)**2 = 8
# D(B, D)**2 = 41
# D(C, D)**2 = 13

#     A    B    C    D
#  A  0
#  B  8    0
#  C  32   8    0
#  D  85   41   13   0
# 2 smallest distances are A and B or B and C
# Among A and B or B and C we are selecting A and B
# D           30.75              0
```

```python
# Step 3: Find and merge the closest clusters
# {A, B}
# New centroid for cluster {A, B} = meanAB = ((1+3)/2, (2+4)/2) = (2, 3)
# (meanAB-meanC)**2 = ((2, 3) - (5, 6))**2 = (4.24)**2 = 18
# Distance to {C}
# D({A, B}, {C}) = Ni*Nj/(Ni+Nj)*(Meani-Meanj)**2 = 2*1/(2+1)*18 = 2/3*18 = 12

# (meanAB-meanD)**2 = ((2, 3) - (8, 8))**2 = (7.81)**2 = 61
# Distance to {D}
# D({A, B}, {D}) = Ni*Nj/(Ni+Nj)*(Meani-Meanj)**2 = 2*1/(2+1)*61 = 2/3*61 = 40.67

#         {A, B}    C          D
# {A, B}  0
# C       12        0
# D       40.67     13         0
# New centroid for {A, B, C} = meanABC = ((1+3+5)/3, (2+4+6)/3) = (3, 4)

# (meanABC-meanD)**2 = ((3, 4) - (8, 8))**2 = (6.40)**2 = 41
# Distance to {D}
# D({A, B, C}, {D}) = Ni*Nj/(Ni+Nj)*(Meani-Meanj)**2 = 3*1/(3+1)*41 = 3/4*41 = 30.75

#             {A, B, C}          D
# {A, B, C}   0
# D           30.75              0
```

**Ward Linkage**

```python
ac3 = AgglomerativeClustering(n_clusters = 5, metric = "euclidean", linkage = "ward")
```

```python
y_hc3 = ac3.fit_predict(x)
```

```python
plt.figure(figsize = (12, 6))

sns.scatterplot(x = x.iloc[y_hc3 == 0,0], y = x.iloc[y_hc3 == 0,1], color = "orange", label = "Cluster One", s = 100)
sns.scatterplot(x = x.iloc[y_hc3 == 1,0], y = x.iloc[y_hc3 == 1,1], color = "red", label = "Cluster Two", s = 100)
sns.scatterplot(x = x.iloc[y_hc3 == 2,0], y = x.iloc[y_hc3 == 2,1], color = "green", label = "Cluster Three", s = 100)
sns.scatterplot(x = x.iloc[y_hc3 == 3,0], y = x.iloc[y_hc3 == 3,1], color = "yellow", label = "Cluster Four", s = 100)
sns.scatterplot(x = x.iloc[y_hc3 == 4,0], y = x.iloc[y_hc3 == 4,1], color = "blue", label = "Cluster Four", s = 100)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Spending Score for Annual Income")
plt.grid()

plt.legend()
plt.show()
```
